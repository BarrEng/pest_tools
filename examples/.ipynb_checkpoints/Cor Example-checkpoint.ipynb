```python
import sys
sys.path.append('../pst_tools')
from matrices import Cor, Cov, Eig

%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':[14,8]})
```

Cor Class is initialized like all the others with a .pst or basename.

```python
cor = Cor('../cc/Columbia')
```

The Cor class has one attribute, df, the parameter correlation coefficient matrix stored in a DataFrame.
The matrix is calucated by pestools and doesn't use the .mtt file from PEST.  This is because often the .mtt file is not writen, and it's just easier this way.


```python
cor.df
```

Cor class currently only has two methods.  plot_heatmap and pars  
plot_heatmap is currently rather simple.  It's just a heatmap of the correlation coefficient matrix  

```python
cor.plot_heatmap(label_rows=False, label_cols=False)
```

Large matrices are hard to deal with.  
The pars method helps with this.  
Just pass it a list of parameters to include.


```python
cor.pars(['kpkpx_tc2', 'kpkpx_tc34', 'kpkpx_ww60', 'kpkpx_ms14', 'kv_w', 'wr_drnc', 'kv_lo', 'sfrc', 'r_col'])
```

You can also pass a list directly to plot_heatmap

```python
cor.plot_heatmap(par_list = ['kpkpx_tc2', 'kpkpx_tc34', 'kpkpx_ww60', 'kpkpx_ms14', 'kv_w', 'wr_drnc', 'kv_lo', 'sfrc', 'r_col'])
```

Seaborn works nice with smaller matrices.  
Plan to add some of the Seaborn options into pestools at some point.  
With large matrix seaborn's extras don't work well

```python
import seaborn

smaller_matrix = cor.pars(['kpkpx_tc2', 'kpkpx_tc34', 'kpkpx_ww60', 'kpkpx_ms14', 'kv_w', 'wr_drnc', 'kv_lo', 'sfrc', 'r_col'])
seaborn.heatmap(smaller_matrix)
```

```python
seaborn.symmatplot(smaller_matrix, names = smaller_matrix.columns.values, cmap = 'RdBu_r')
```
