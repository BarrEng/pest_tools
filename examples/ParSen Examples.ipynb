# Examples of working wtih ParSen Class

ParSen class works similar to the Res class to initiate it.  It only needs the PEST base name, or Pest control file.
It uses the base name to get the observations and parameters from the PEST control file and read the .jco file.
ParSen class does not use the sensitivity output from PEST.  It calculates sensitivies itself so that observations can
be excluded to allow several layers of analysis

```python
parsen = ParSen('../cc/Columbia')
```

### Plot the 30 most sensitive parameters

```python
parsen.plot(n = 30)
```

### Plot the most sensitive parameter for a single group

```python
parsen.plot(n=30, group = 'kp')
```

### Alternative lables can be defined with a dictionary.

```python
alt_labels = {'r_col' : 'Full Parameter Description1', 'kh_cr' : 'K of the something'}
parsen.plot(n = 30, alt_labels = alt_labels)
```

### Plot the sum of sensitivity by group

```python
parsen.plot_sum_group()
```

### Use alternative group names and change the color

```python
alt_labels = {'sfr_cond' : 'Stream Conductance', 'rech' : 'Recharge', 'kz' : 'Vertical K', 'Kp' : 'Hydraulic Conductivity'}
parsen.plot_sum_group(alt_labels = alt_labels, color = "DarkGray")
```

### Plot the mean of sensitivity of group

```python
parsen.plot_mean_group(alt_labels = alt_labels, color = 'LightGreen', linewidth = 0.0)
```

### plot() method uses the 'Set2' cmap by default to get colors for different groups.  A different cmap can be used

```python
parsen.plot(n=30, cmap = 'jet')
```

### You can also use a dictionary with group names yp set colors for groups

```python
color_dict = {'kz': (0.89411765336990356, 0.10196078568696976, 0.10980392247438431, 1.0), 'kp': 'g', 'sfr_cond' : 'b',
              'rech' : 'k'}
parsen.plot(n=20, color_dict = color_dict)
```

# Exploring more

### When initiating the ParSen class there a a number of options that allow more detailed exploration of the sensitivities

###### Exclude the regularization observations

```python
parsen = ParSen('../cc/Columbia', drop_regul = True)
parsen.plot(n=20)
```

###### exclude observation groups

```python
parsen = ParSen('../cc/Columbia', drop_groups = ['head_poor', 'WCRs1'], drop_regul = False)
parsen.plot(n=20)
```

###### only keep one or a few groups

```python
parsen = ParSen('../cc/Columbia', keep_groups = ['cc_streams'])
parsen.plot(n=20)
```

###### only keep a single observation

```python
parsen = ParSen('../cc/Columbia', keep_obs = ['7088563401_g'])
parsen.plot(n=20)
```
