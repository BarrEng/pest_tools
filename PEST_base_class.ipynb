###base class for PEST run:
* reads run information from PEST control file
* goal is to have something that can be inherited by other classes, similar to flopy base classes
* could also add write method, which would write out a new PEST control file

```python
import os
import numpy as np
import pandas as pd

class PESTbase(object):
    """
    Base class for PEST runs
    """
    
    def __init__(self, basename, run_folder=None):
        
        self.__name = basename
        self.pstfile = self.__name + '.pst'
        if run_folder is None: run_folder = os.getcwd()
        self.run_folder = run_folder
        
        # PEST variables (should there be a separate object for these?)
        self.RSTFLE = 'restart'
        self.PESTMODE = 'estimation'

        # control data
        self.NPAR = 1
        self.NOBS = 1
        self.NPARGP = 1
        self.NPRIOR = 0
        self.NOBSGP = 1
        
        self.NTPLFLE = 1
        self.NINSFLE = 1
        self.PRECIS = 'double'
        self.DPOINT = 'point'
        
        self.RLAMBDA1 = 20
        self.RLAMFAC = -3
        self.PHIRATSUF = 0.3
        self.PHIREDLAM = 0.01
        self.NUMLAM = -10
        self.JACUPDATE = 999
        self.LAMFORGIVE = 'lamforgive'
        self.DERFORGIVE = ''
        
        self.RELPARMAX = 10
        self.FACPARMAX = 10
        self.FACORIG = 0.001
        
        self.PHIREDSWH = 0.1
        self.NOPTSWITCH = 1
        self.DOAUI = 'noaui'
        
        self.NOPTMAX = 0
        self.PHIREDSTP = 0.01
        self.NPHISTP = 3
        self.NPHINORED = 3
        self.RELPARSTP = 0.01
        self.NRELPAR = 3
        
        self.ICOV = 0
        self.ICOR = 0
        self.IEIG = 0
        self.REISAVEITN = 'REISAVEITN'
        self.PARSAVEITN = 'PARSAVEITN'
        
        # singular value decomposition
        self.SVD = False
        self.SVDMODE = 1,
        self.MAXSING = 1
        self.EIGTHRESH = 5e-7
        self.EIGWRITE = 1
        
        # svd assist
        self.svda = {}
        
        # parameter groups
        self.pargroups = {}
        self.pargroups_attr = ['PARGPNME', 'INCTYP', 'DERINC', 'DERINCLB', 'FORCEN', 
                               'DERINCMUL', 'DERMTHD']

        # parameter data
        self.paradata = pd.DataFrame()
        
        # observation groups
        self.obsgroups = []
        
        # observation data
        self.obsdata = pd.DataFrame()
        
        # model command line
        self.batchfile = ''
        
        # model input/output
        self.io = ''
        
        # prior information
        self.prior = []
        
        # regularisation
        self.PHIMLIM = self.NOBS
        self.PHIMACCEPT = self.PHIMLIM * 1.05
        self.FRACPHIM = 0.10
        self.WFINIT = 1.0
        self.WFMIN = 1e-10
        self.WFMAX = 1e10
        self.REGCONTINUE = ''
        self.WFFAC = 1.3
        self.WFTOL = 0.01
        self.IREGADJ = 1
        
        
    def read_pst(self):
        """
        Read run information from the PEST control file
        """
        
        f = os.path.join(self.run_folder, self.pstfile)
        print 'reading {}...'.format(f)
        
        pst = open(f).readlines()
        
        self.RSTFLE, self.PESTMODE = pst[2].strip().split()

        print 'control data...'
        self.NPAR, self.NOBS, self.NPARGP, self.NPRIOR, self.NOBSGP \
        = [int(s) for s in pst[3].strip().split()[0:5]]
        
        self.NTPLFLE, self.NINSFLE = [int(s) for s in pst[4].strip().split()[0:2]]
        self.PRECIS, self.DPOINT = pst[4].strip().split()[2:4]
        
        self.RLAMBDA1, self.RLAMFAC, self.PHIRATSUF, self.PHIREDLAM = [float(s) for s in pst[5].strip().split()[0:4]]
        self.NUMLAM = int(pst[5].strip().split()[4])
        try:
            self.JACUPDATE = int(pst[5].strip().split()[5])
            self.LAMFORGIVE = pst[5].strip().split()[6]
            self.DERFORGIVE = pst[5].strip().split()[7]
        except:
            pass
        
        self.RELPARMAX, self.FACPARMAX, self.FACORIG = [float(s) for s in pst[6].strip().split()[0:3]]
        
        self.PHIREDSWH = float(pst[7].strip().split()[0])
        try:
            self.NOPTSWITCH = int(pst[7].strip().split()[1])
        except:
            pass
        try:
            self.DOAUI = [s for s in pst[7].strip().split() if 'aui' in s][0]
        except:
            pass
        
        self.NOPTMAX = int(pst[8].strip().split()[0])
        self.PHIREDSTP = float(pst[8].strip().split()[1])
        self.NPHISTP = int(pst[8].strip().split()[2])
        self.NPHINORED = int(pst[8].strip().split()[3])
        self.RELPARSTP = float(pst[8].strip().split()[4])
        self.NRELPAR = int(pst[8].strip().split()[5])
        
        self.ICOV, self.ICOR, self.IEIG = [int(s) for s in pst[9].strip().split()[0:3]]
        
        if not 'reisaveitn' in pst[9].lower():
            self.REISAVEITN = ''
        if not 'parsaveitn' in pst[9].lower():
            self.PARSAVEITN = ''
            
        # read rest of pst file
        knt = 10
        for line in pst[10:]:
            
            if 'singular value decompostion' in line:
                print 'singular value decompostion...'
                self.SVD = True
                self.SVDMODE = int(pst[knt + 1].strip().split()[0])
                self.MAXSING = int(pst[knt + 2].strip().split()[0])
                self.EIGTHRESH = float(pst[knt + 2].strip().split()[1])
                self.EIGWRITE = int(pst[knt + 3].strip().split()[0])
                
            if 'parameter groups' in line:
                print 'parameter groups...'
                for i in np.arange(self.NPARGP) + 1:
                    self.read_par_group(pst[knt + i])
                    
            knt += 1
            
    def read_par_group(self, l):
        """
        convience function to read par group information into a dictionary
        """
        l = l.strip().split()
        pargp = [l[0], l[1], float(l[2]), float(l[3]), l[4], float(l[5]), l[6]]
        
        self.pargroups[pargp[0]] = dict(zip(self.pargroups_attr, pargp))
        
```

```python
cpest = PESTbase('columbia_opt5_13best', run_folder='cc')
cpest.read_pst()
```
