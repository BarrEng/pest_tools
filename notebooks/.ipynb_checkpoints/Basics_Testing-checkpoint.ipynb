```python
import pest_tools as pt
import numpy as np
%matplotlib inline
```

All classes build off the Pst class.  All ask for the path to the PEST control file at a minimum.
For reading in files such as the .jco or .res file the base name in the PEST control file is used.

```python
# Path to PEST control file
# Assumes that it is in the same directory as this notebook
pest_ctl = 'example.pst'
```

# Pst Class

Pst class is the base or "super" class.  It has attributes such as:
* ob_groups
* obs
* obs_df
* par_groups
* pars
* pars_df

These attributes are used by other classes.
 Pst class also hold some "private" methods and attributes that are used by other classes but not intended for typical use by the user:
 * _load_jco
 * _obs_dict
 * _pars_dict

```python
pst = pt.Pst(pest_ctl)
```

```python
# Observation groups
pst.ob_groups
```

```python
# Parameters DataFrame
pst.pars_df
```

# Res Class

The Res class is for items dealing with residuals
Attributes:
* res_df : Pandas Data Frame

Methods:
* plot_measure_vs_model()
* plot_measured_vs_residual()
* plot_objective_contrib()
* objective_contrib()

The Res class requires just the path to the Pest Control file.  Optional argument of 'res' or 'rei' but that neeeds some work

```python
res = pt.Res(pest_ctl)
```

## Residual Plots

```python
res.plot_measure_vs_model(groups = ['sny08','cwi'], plot_type = 'hexbin')
```

```python
res.plot_objective_contrib()
```

Attributes from the Pst class are available in the Res class also

```python
res.ob_groups
```

# ParSen Class

ParSen class is similar in behavior to the Res class but deals with parameter sensitivity

Attributes:
* par_sen_df : Pandas DataFrame 

            
Methods
* plot()
* tail()
* head()
* par()
* group()
* sum_group()
* plot_sum_group()
* plot_mean_group()

```python
par_sen = pt.ParSen(pest_ctl)
```

```python
par_sen.head(20)
```

```python
par_sen.plot(n=30)
```

```python
par_sen.plot(group = 'kx', n = 20)
```
